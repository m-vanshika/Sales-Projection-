```python
#ca;culation
import numpy as np
#import and transform data
import pandas as pd
#visualization
import matplotlib.pyplot as plt
#to generate regression model using linear regression
from sklearn import linear_model
#data visusalisation
import seaborn as sns
#plotting multivectore
import mpl_toolkits
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#data frame
df=pd.read_excel("Linear Regression Data.xlsx")
```

```python
df
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
#features have a corellation to itself
#even with each other they have a high corelation
#Therefore change in one variable affects change in the other
```

```python

plt.figure(figsize=(12,16))
#drawing a scatter plot
plt.scatter(df['Marketing Budget (X) in Thousands'],df['Actual Sales(Y) in Millions'],color='red')
plt.title("Sales vs Budget",fontsize=24)
plt.xlabel("Marketing budget")
plt.ylabel("Actual Sales")
plt.grid(True)
plt.show()
```

```python
n_df=df.drop('Actual Sales(Y) in Millions',axis=1)
n_df
```

```python
sales=df['Actual Sales(Y) in Millions']
sales
```

```python
#this is the object of linear regression
lr=linear_model.LinearRegression()
lr.fit(n_df,sales)
```

```python
lr.predict([[150]])
```

```python
y_pred=lr.predict(n_df)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['Marketing Budget (X) in Thousands'],df['Actual Sales(Y) in Millions'],color='red')
plt.plot(df['Marketing Budget (X) in Thousands'],y_pred,'y')
plt.show()
```

```python
plt.figure(figsize=(10,10))
x_ax=range(len(df['Actual Sales(Y) in Millions']))
plt.plot(x_ax,df['Actual Sales(Y) in Millions'],label='original')
plt.plot(x_ax,y_pred,label='Predicted')
plt.title("Comparision")
plt.legend()
plt.show()
```
